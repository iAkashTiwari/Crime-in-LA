```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Crime_Data_from_2020_to_Present.csv')
df
```

# =============== Data Exploration And Data Cleaning ===============

```python
df.shape
```

```python
df.drop(['DR_NO', 'Date Rptd', 'Rpt Dist No' , 'Part 1-2', 'Mocodes', 'Crm Cd 1', 'Crm Cd 2', 'Crm Cd 3', 'Crm Cd 4', 'Cross Street', 'LOCATION'], axis=1, inplace=True)
df
```

```python
# Check whether dataset contains any Null or NaN values
df.isnull().sum()
```

```python
# Drop rows contains missing values
df.dropna(axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
df.isnull().sum()
```

# =============== Crime Analysis ===============

```python
# Crime by Area
sns.countplot(data=df, x='AREA NAME', order=df['AREA NAME'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Crime Distribution by Area in Los Angeles')
plt.xlabel('Area Name')
plt.ylabel('Crime Count')
plt.show()
```

```python
# Crime by Type
sns.countplot(data=df, y='Crm Cd Desc', order=df['Crm Cd Desc'].value_counts().index[:10])
plt.title('Top 10 Most Common Crimes in Los Angeles')
plt.xlabel('Count')
plt.ylabel('Crime Type')
plt.show()
```

```python
# Crime by Victim Age
sns.histplot(data=df, x='Vict Age', bins=20)
plt.title('Distribution of Victim Ages in Crimes')
plt.xlabel('Victim Age')
plt.ylabel('Count')
plt.show()
```

```python
# Convert 'TIME OCC' to a proper time format
df['TIME OCC'] = df['TIME OCC'].apply(lambda x: f"{int(x):04d}")
df['TIME OCC'] = pd.to_datetime(df['TIME OCC'], format='%H%M').dt.strftime('%H:%M:%S')
```

```python
# Extract the hour from the datetime column
df['Hour'] = pd.to_datetime(df['TIME OCC'], format='%H:%M:%S').dt.hour
```

```python
# Visualize crime by time of day
sns.countplot(data=df, x='Hour', order=df['Hour'].value_counts().index)
plt.title('Crime Distribution by Time of Day in Los Angeles')
plt.xlabel('Hour of Day')
plt.ylabel('Crime Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Visualize crime by victim descent
sns.countplot(data=df, x='Vict Descent', order=df['Vict Descent'].value_counts().index)
plt.title('Crime Distribution by Victim Descent in Los Angeles')
plt.xlabel('Victim Descent')
plt.ylabel('Crime Count')
plt.xticks(rotation=90)
plt.show()
```
